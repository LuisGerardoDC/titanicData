## Porcentaje sobrevivieron y murieron 

```python
df_original['Survived']= df_original['Survived'].apply(lambda x: 'Survived' if (x == 1) else 'Died')
df_original['Survived'] = df_original['Survived'].astype("category")
```

```python
df_original.groupby(['Survived'])['Survived'].count().apply(lambda x: (x*100)/df_original.shape[0])
```

## Personas que sobrevivieron por clase

```python
(df_original
 .groupby(['Pclass','Survived'])['Survived']
 .count()
 .apply(lambda x: (x*100)/df_original.shape[0]).to_frame() 
)
```

## eliminar los usuarios con edad desconocida

```python
df_edda_filtrada=  df_original[~df_original['Age'].isna()]
df_edda_filtrada
```

## edad media de las mujeres por clase

```python
df_mujeres = df_edda_filtrada[df_edda_filtrada['Sex'] == 'female']
df_mujeres.groupby(['Pclass'])['Age'].mean().to_frame()
```

## Columna booleana menor de edad

```python
df_menores = df_edda_filtrada.copy()
```

```python
df_menores['adult']=df_menores['Age'].apply(lambda x: False if (x < 18) else True)
```

## Porcentaje menores y mayores de edad

```python
(df_menores
 .groupby(['adult'])['adult']
 .count()
 .apply(lambda x: (x*100)/df_menores.shape[0])
 .to_frame()
)
```
